## Histograms

* Displays a distribution of the population

```python
T2.hist('Y')
T2
```

* Histograms are left bound inclusive, right bound exclusive (think like np.arange)

* Total area sums to 1, or 100%.
* The area of each bar represents the proportion represented in that bin.
* Formula for percent representation: y (percent) * x (bin width)
* Formula for proprotion representation: y (proportion) * x (bin width)

```python
T2.group('X').hist('count', bins = [.5, 1.5, 2.5, 3.5])
T2.show()
```
